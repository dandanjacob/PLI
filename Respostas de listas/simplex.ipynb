```julia
using Pkg
Pkg.add("JuMP")
Pkg.add("GLPK")
```

```julia
using JuMP
using GLPK

# Defina o modelo
model = Model(GLPK.Optimizer)

# Defina as variáveis
@variable(model, x1 >= 0)
@variable(model, x2 >= 0)
@variable(model, x3 >= 0)

# Defina a função objetivo
@objective(model, Min, -10x1 - 12x2 - 12x3)

# Adicione as restrições
@constraint(model, x1 + 2x2 + 2x3 <= 20)
@constraint(model, 2x1 + x2 + 2x3 <= 20)
@constraint(model, 2x1 + 2x2 + x3 <= 20)

# Resolver o problema
optimize!(model)

# Obter e imprimir a solução
optimal_x1 = value(x1)
optimal_x2 = value(x2)
optimal_x3 = value(x3)
optimal_value = objective_value(model)

println("Valor ótimo de x1: ", optimal_x1)
println("Valor ótimo de x2: ", optimal_x2)
println("Valor ótimo de x3: ", optimal_x3)
println("Valor ótimo da função objetivo: ", optimal_value)
```
